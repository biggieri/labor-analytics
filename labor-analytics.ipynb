```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
random.seed(42)
np.random.seed(42)

# Define departments and employees
departments = {
    "Operations": ["Carlos M", "Diana R", "James T", "Sofia L", "Mark H"],
    "Warehouse": ["Pedro G", "Anna K", "Luis F", "Rachel B", "Tom W"],
    "Delivery": ["Miguel A", "Sara N", "Kevin P", "Maria C", "David S"],
    "Customer Service": ["Emma J", "Chris V", "Nina O", "Paul E", "Laura Z"]
}

hourly_rates = {
    "Operations": 22,
    "Warehouse": 18,
    "Delivery": 20,
    "Customer Service": 16
}

# Generate 750 rows of labor data
rows = []
start_date = datetime(2025, 1, 1)

for _ in range(750):
    dept = random.choice(list(departments.keys()))
    employee = random.choice(departments[dept])
    hours = round(random.uniform(4, 10), 1)
    rate = hourly_rates[dept]
    date = start_date + timedelta(days=random.randint(0, 364))
    
    rows.append({
        "date": date.strftime("%Y-%m-%d"),
        "employee": employee,
        "department": dept,
        "hours_worked": hours,
        "hourly_rate": rate,
        "labor_cost": round(hours * rate, 2)
    })

df = pd.DataFrame(rows)
df["date"] = pd.to_datetime(df["date"])
df["month"] = df["date"].dt.to_period("M")

print(f"Dataset shape: {df.shape}")
df.head(10)
```

```python
working_df = df.copy()

working_df.isnull().sum()

working_df["employee"] = working_df["employee"].str.title()
working_df["department"] = working_df["department"].str.title()

print("Duplicates found:")
print(working_df[working_df.duplicated(subset=["date","employee"], keep=False)])
```

```python
# Remove duplicates keeping first occurrence
working_df = working_df.drop_duplicates(subset=["date", "employee"], keep="first")

print(f"Rows before: 750")
print(f"Rows after: {len(working_df)}")
print(f"Duplicates removed: {750 - len(working_df)}")
```
